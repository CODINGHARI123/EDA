```python
# Define numeric columns
numeric_columns = ['Inches', 'Ram(GB)', 'Weight(KG)', 'Price']

# Function to identify and impute outliers
def impute_outliers_with_median(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define outliers as those below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Calculate median of the column
    median_value = df[column].median()
    
    # Impute outliers with the mean
    df[column] = df[column].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

# Impute outliers for each numeric column and plot boxplots
for column in numeric_columns:
    # Impute outliers
    impute_outliers_with_median(df, column)
    
    # Plot boxplot
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, y=column)
    plt.title(f'Boxplot of {column} After Imputation')
    plt.xlabel(column)
    plt.show()
```
